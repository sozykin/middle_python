```python
import requests
from bs4 import BeautifulSoup
import json
```

## Загружаем страницу

```python
url = 'https://hh.ru/search/vacancy'
user_agent = {'User-agent': 'Mozilla/5.0'}
```

```python
url_params = {
    "text": "middle python",
    "search_field": "name",
    "per_page": "30",
    "page": "2"
}
```

```python
result = requests.get(url, headers=user_agent, params=url_params)
```

```python
result.url
```

```python
result.status_code
```

```python
result.content.decode()
```

## Парсинг страницы



```python
# Создаем объект soup на основе загруженной Web-страницы
soup = BeautifulSoup(result.content.decode(), 'lxml')
```

## Список вакансий

```python
vacancies = soup.find_all('a', attrs={'data-qa': 'serp-item__title'})
```

```python
for vacancy in vacancies:
    print(vacancy.text, vacancy.get('href'))
```
